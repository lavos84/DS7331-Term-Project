<br>**Borough**  The borough that the tax lot is located in

<br>**Block**  The tax block that the tax lot is located in
<br>**Lot**  The number of the tax lot
<br>**Address**  An address for the tax lot
<br>**Parcel Name**  General Property name

<br>**Structure Completed**  The year construction of the structure(s) was completed and whether such year is an estimate, where applicable
<br>**Number Structures**  The number of structures
<br>**Total Gross Area Structures**  Total gross area of all structures expressed in square feet rounded to the nearest integer, where applicable
<br>**Ratio Building to Floor Area**  Ratio of building floor area to the area of the property, where applicable
<br>**Allowable Building to Floor Area**  Allowable ratio of building floor area to the area of the property, where applicable
<br>**Land Use Category**  A code for the tax lot's land use category, modified for display of parks, New York City Department of Parks and Recreation properties and New York State Office of Parks, Recreation and Historic Preservation properties in the appropriate category on land
<br>**Community Board**  The community district (CD) or joint interest area (JIA) that the tax lot is located in, or partially located in
<br>**Census Tract**  The 2010 census tract that the tax lot is located in
<br>**Census Block**  The 2010 census block that the tax lot is located in
<br>**School Dist**  The community school district that the tax lot is located in
<br>**Council District**  The city council district that the tax lot is located in
<br>**Postcode**  The zip code that the tax lot is located in
<br>**Fire Comp**  The fire company that services the property
<br>**Health Area**  The health area
<br>**Health Ctr**  The health center district
<br>**Police Prct**  The police precinct
<br>**Major Use**  The major use of the structure or structures, where applicable
<br>**Number of Easements**  The number of easements, where applicable
<br>**Commercial Floor Area**  The exterior dimensions of the portion of the structure or structures allocated for commercial use, where applicable
<br>**Residential Floor Area**  The exterior dimensions of the portion of the structure or structures allocated for residential use, where applicable
<br>**Office Floor Area**  An estimate of the exterior dimensions of the portion of the structure(s) allocated for office use
<br>**Retail Floor Area**  An estimate of the exterior dimensions of the portion of the structure(s) allocated for retail use
<br>**Garage Floor Area**  An estimate of the exterior dimensions of the portion of the structure(s) allocated for garage use
<br>**Storage Floor Area**  An estimate of the exterior dimensions of the portion of the structure(s) allocated for storage or loft purposes
<br>**Factory Floor Area**  An estimate of the exterior dimensions of the portion of the structure(s) allocated for factory, warehouse or loft use
<br>**Other Floor Area**  An estimate of the exterior dimensions of the portion of the structure(s) allocated for other than Residential, Office, Retail, Garage, Storage, Loft or Factory use
<br>**Num Floors**  In the primary building on the tax lot, the number of full and partial stories starting from the ground floor
<br>**Residential Units**  The sum of residential units in all buildings on the tax lot
<br>**Residential and Non-Residential Units**  The sum of residential and non-residential units in all structures, where applicable
<br>**Lot Front**  The frontage, measured in feet
<br>**Lot Depth**  The depth, measured in feet
<br>**Bldg Front**  The frontage along the street, measured in feet
<br>**Bldg Depth**  The depth of the structure or structures, which is the effective perpendicular distance, measured in feet, where applicable
<br>**Proximity Code**  Whether the structure or structures are detached, semi-detached or attached to neighboring structures, where applicable
<br>**Irr Lot Code**  Whether the property is irregularly shaped
<br>**Lot Type Code**  The location relative to another lot or water, expressed as mixed or unknown, block assemblage, waterfront, corner, through, inside, interior, island, alley, or submerged land
<br>**Bsmt Code**  A code describing the basement
<br>**Assess Land**  The actual assessed land value for Fiscal Year 2013
<br>**Exempt Land**  The actual exempt land value, which is determined differently for each exemption program, is the dollar amount related to that portion of the tax lot that has received an exemption or abatement for Fiscal Year 2013
<br>**Exempt Tot**  The actual exempt total value, which is determined differently for each exemption program, is the dollar amount related to that portion of the tax lot that has received an exemption or abatement for Fiscal Year 2013
<br>**Year Alter 1**  If a building has only been altered once, this is the date of the alteration
<br>**Year Alter 2**  If a building has been altered more than once, this is the year of the most recent alteration
<br>**His Dist**  The name of the Historic District as designated by the New York City Landmarks Preservation Commission
<br>**Landmark**  The name of an individual landmark, landmark site (e.g. Richmondtown Restoration), or an interior landmark, as designated by the New York City Landmarks Preservation Commission
<br>**Condominium Number**  The condominium number assigned to the complex, where applicable
<br>**Coordinates**  The coordinate of the XY coordinate pair that depicts the property’s approximate location as expressed in the New York-Long Island State Plane coordination system
<br>**E-Designation Number**  The e-designation number associated with the property, where applicable
<br>**Industrial Business Zone**  Whether the property is located in an industrial business zone
<br>**Zone Dist 1**  The primary zoning classification of the property
<br>**Zone Dist 2**  The zoning designation occupying the second greatest percentage of the property’s area, where applicable
<br>**Overlay 1**  The primary commercial overlay assigned to the property, where applicable
<br>**Overlay 2**  The commercial overlay occupying the second greatest percentage of the property’s area, where applicable
<br>**SP Dist 1**  The special purpose district assigned to the tax lot
<br>**SP Dist 2**  The special purpose district assigned to the tax lot
<br>**Potential Urban Ag**  Whether the land is suitable for urban agriculture (for description, click “About” then “Attachments”)
<br>**Contact**  Agency contact information, including name, telephone number and email address
