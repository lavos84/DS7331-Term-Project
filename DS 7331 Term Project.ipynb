#### Plot the density of missing values

To do this we'll first count the number of NA's (NaN) in each column and convert that to a percentage.

```python
nas = (nyc.isna().sum()).sort_values()
na_percent = (nas / (nas + nyc.count()) 
              * 100).sort_values()
colnames = list(nas.index)
```

Then, using matplotlib, we can create a horizontal bar chart with each column on the vertical axis, and the percentage of missing values on the horizontal axis.  For style, we'll also annotate the number of NA's (count) to the right of each horizontal bar.

```python
%matplotlib inline

plt.style.use('ggplot')
# This section of code was inspired by:
# https://stackoverflow.com/questions/28931224/adding-value-labels-on-a-matplotlib-bar-chart

fig = plt.figure(figsize = (15, 15))
ax = na_percent.plot(kind ='barh')
ax.set_title('Missing Value Density', fontsize = 20)
ax.set_xlabel('Percent Missing (%)', fontsize = 18)
ax.set_yticklabels(colnames)

def add_counts(ax, nas, spacing=15):
    # For each bar, place a label
    for rect, count in zip(ax.patches, nas):
        # Get X and Y placement of label from rect.
        x_value = rect.get_width()
        y_value = rect.get_y() 
        - rect.get_height() / 2    # This makes the text appear in the middle of the bar

        space = spacing  # Distance to the right the annotation appears
        label = count

        # Create the annotation
        ax.annotate(label, 
                    (x_value, y_value), 
                    xytext=(space, 0), 
                    textcoords = "offset points", 
                    ha = 'center', 
                    va = 'bottom')

# Add the counts
add_counts(ax, nas)
# Add a verticle line at 90%
ax.axvline(x = 90, color = "red", alpha = .5)
```

##### Generate missing BINs

To remedy this situation, for every missing BIN we can multiply its Borough number by 1,000,000 to generate the correct BIN for that property.  The Missing Value Density plot reveals that the Borough feature has no missing values, so we will convert the Borough value from a catagorical factor to an integer before multiplying by 1,000,000.

```python
# Multiply Borough (as an integer) by 1,000,000 at the identified indices and assign it BIN
nyc.loc[nyc.BIN.isna() , "BIN"] = 1000000 * nyc.loc[nyc.BIN.isna(), "Borough"].astype(int)
```

#### Separating the "Current Uses" column

The "Current Uses" column is interesting in that it lists the known uses for each building, and would be a good metric to predict or cluster.  What makes it interesting though is how the uses are listed in the data set.  For most records, different uses are separated by a comma and a space.  However, there are several instances where uses are separated by a semicolon, and others still that are separated by a semicolon and a space.  We will need consistency to make use of this feature.

To accomplish this, we will replace the semicolon (and the semicolon with a space) with a comma and a space.  Then, each different use can be identified by splitting the entry at each comma and space.

```python
# -------- I don't know how to format this to PEP 8----------

# Replace the ";"s and "; "s with ", " everywhere data exists,
# then split by ", "s.
nyc.loc[nyc["Current Uses"].notna(), "Current Uses"] = nyc.loc[nyc["Current Uses"].notna(), "Current Uses"].str.replace("; ; ", ", ").str.replace("; ", ", ").str.replace(";", ", ").str.split(", ")
```

Now that the "Current Uses" are stored as a list for each record, we should identify how many unique values there are and if any are similar enough to merit being grouped into a single entity.

But first, we must address the missing values in this column.  Where there is a missing value, we will replace it with a "NO DATA"

```python
# Replace missing values with "NO DATA"
nyc.loc[nyc["Current Uses"].isna(), "Current Uses"] = ["NO DATA"]

# Identify every unique "Current Use"
unique_uses = []
for record in nyc["Current Uses"]:
    if record != "NO DATA":
        for use in record:
            if use not in unique_uses:
                unique_uses.append(use)

unique_uses.sort()
```

#### Visualize the frequency of uses

We have the various uses for each building.  Now let's see what the distribution looks like.

```python
uses = pd.DataFrame({"Counts": np.zeros(len(unique_uses))}, index = [unique_uses])
for record in nyc["Current Uses"]:
    if record != "NO DATA":
        for use in record:
            uses.loc[use, "Counts"] += 1
```

```python
plt.style.use('ggplot')
# This section of code was inspired by:
# https://stackoverflow.com/questions/28931224/adding-value-labels-on-a-matplotlib-bar-chart

fig = plt.figure(figsize = (30, 30))
ax = uses.Counts.plot(kind ='bar')
ax.set_title('Histogram of Current Uses', fontsize = 30)
ax.set_xticklabels(unique_uses, fontsize = 10)
ax.set_ylabel("Frequency", fontsize = 20)

def add_counts2(ax, counts, spacing=15):
    # For each bar, place a label
    for rect, count in zip(ax.patches, counts):
        # Get X and Y placement of label from rect.
        x_value = rect.get_x() + rect.get_width() / 2
        y_value = rect.get_height()

        space = spacing  # Distance to the right the annotation appears
        label = int(count)

        # Create the annotation
        ax.annotate(label, 
                    (x_value, y_value), 
                    xytext=(0, space), 
                    textcoords = "offset points", 
                    ha = 'center', 
                    va = 'bottom',
                    fontsize = 12,
                    rotation = 90)

# Add the counts
add_counts2(ax, uses.Counts)
```

#### Assessed Land Value

To gauge the differences in assessed land values we will find the total assessed land values in each district and color code it on the plot above based on that sum value.  The district is given by the "Community Board" feature in our data.  We will group the data by Community Board and average the assessed land values where it is known.

```python
# Collect data on the Assessed Land value by Community
mean_land_value = pd.DataFrame(nyc.groupby(["Community Board"])["Assess Land"].agg("mean"))
median_land_value = pd.DataFrame(nyc.groupby(["Community Board"])["Assess Land"].agg("median"))
max_land_value = pd.DataFrame(nyc.groupby(["Community Board"])["Assess Land"].agg("max"))
total_land_value = pd.DataFrame(nyc.groupby(["Community Board"])["Assess Land"].agg("sum"))
# Convert the index to a column of integer type
mean_land_value["BoroCD"] = mean_land_value.index.astype(int)
median_land_value["BoroCD"] = median_land_value.index.astype(int)
max_land_value["BoroCD"] = max_land_value.index.astype(int)
total_land_value["BoroCD"] = total_land_value.index.astype(int)
# Change the column name for clarification
mean_land_value = mean_land_value.rename(columns = {"Assess Land": "Mean_value"})
median_land_value = median_land_value.rename(columns = {"Assess Land": "Median_value"})
max_land_value = max_land_value.rename(columns = {"Assess Land": "Max_value"})
total_land_value = total_land_value.rename(columns = {"Assess Land": "Total_value"})
```

```python
# Merge the data to the districts dataframe
districts = districts.merge(mean_land_value, on = "BoroCD")
districts = districts.merge(median_land_value, on = "BoroCD")
districts = districts.merge(max_land_value, on = "BoroCD")
districts = districts.merge(total_land_value, on = "BoroCD")
```

```python
# Plot the data on the map of New York City
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, ax = plt.subplots(figsize = (15,15))
ax.set_aspect('equal')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
boroughs.plot(ax=ax, column = "BoroName", cmap = "OrRd", edgecolor = "black")
# Overlay the districts that make up each borough
districts.plot(ax=ax, column = "Total_value", cmap = "Spectral_r", edgecolor = "black", legend = True, cax = cax)
plt.show()
```

Whether we're looking at mean, median, max, or total assessed land values, there seems to be a consistent trend in the data centered around districts 6 (orange in the center of Manhattan), 28 and 31 (maroon in the lower portion of Queens).  Upon further investigation, district 6 is home to Central Park.  Intuitively it makes sense that land values would be higher there given the location.  Districts 28 and 31 together make up John F. Kennedy International Airport.  Again, given the significance of these locations, the extreme assessed land values found here are unsurprising.
